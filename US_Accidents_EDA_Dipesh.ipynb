###  Start Time

```python
# Time stamping for date as well as time in data frame 
df.Start_Time = pd.to_datetime(df.Start_Time)
df.Start_Time[0]
```

```python
## figure out how to show percentages 
sns.distplot(df.Start_Time.dt.hour,bins =24, kde = False, norm_hist = True)
```

```python
# Figuring out cities with accidents according to weeks and plotting
# histogram plot 
sns.distplot(df.Start_Time.dt.dayofweek,bins =24, kde = False, norm_hist = True)
```

```python
# looking for trend of accidents on monday using time stamp and plotting hist plot 
mondays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(mondays_start_time.dt.hour, bins=24, kde=False, norm_hist= True)
```

```python
# Plotting graph for month wise accidents 
sns.distplot(mondays_start_time.dt.month, bins=12, kde=False, norm_hist= True)
```

```python
df_2018 = df[df.Start_Time.dt.year == 2018]
sns.distplot(df_2018.Start_Time.dt.month, bins= 12, kde =False, norm_hist= True)
```
